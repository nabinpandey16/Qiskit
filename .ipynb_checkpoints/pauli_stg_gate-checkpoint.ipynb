```python
from qiskit import*
from qiskit.visualization import plot_bloch_multivector, visualize_transition, plot_histogram
from qiskit import QuantumCircuit
import numpy as np
import matplotlib.pyplot as plt
from qiskit_aer import Aer
```

```python
qc=QuantumCircuit(1,1)
```

```python
# initial_state =[0,1]
# qc.initialize(initial_state,0)
```

```python
# qc.x(0) #we are adding this x gate, so  we can visualize bloch sphere better
```

```python
qc.h(0) #applying h gate 
qc.measure(0,0) #here we measure and apply a classical bit and we must have to add a classical bit on qc as QuantumCircuit(1,1)
# qc.z(0) #applying z gate 
# qc.h(0) #applying h gate 
```

```python
qc.draw('mpl')
```

```python
backend=Aer.get_backend('statevector_simulator')
```

```python
out=backend.run(qc)
```

```python
statevector = out.result().get_statevector()
```

```python
#visualization of bloch sphere 
plot_bloch_multivector(statevector)
```
