```python
#create a quantum circuit with 3 qubits
qc=QuantumCircuit(2)
```

```python
# initial_state =[0,1]
# qc.initialize(initial_state,0)
```

```python
#Applying H gate on controlled qubit
qc.h(0)
```

```python
#Applying CNOT gate
qc.cx(0,1)
```

```python
qc.draw('mpl')
```

```python
backend=Aer.get_backend('statevector_simulator')
```
