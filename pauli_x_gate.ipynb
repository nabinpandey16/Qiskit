```python
#importing qiskit
from qiskit import*
 #print(qiskit.__version__)
```

```python
from qiskit.visualization import plot_bloch_multivector
```

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
```

```python
qc=QuantumCircuit(1)
```

```python
#Applying Pauli's X-Gate
qc.x(0)
```

```python
#Visualizing the X-gate
qc.draw('mpl')
```

```python
from qiskit_aer import Aer
```

```python
backend=Aer.get_backend('statevector_simulator')
```

```python
out=backend.run(qc)
```

```python
statevector = out.result().get_statevector()
```

```python
#vidualization of bloch sphere 
plot_bloch_multivector(statevector)
```
